```python
#─ Cell 2：進主頁 + 取得行政區列表 ───────────────────────
def setup_and_get_districts():
    """
    進主頁 → 進 map → 進 query 頁
    同時取得：current_csrf、captchaKey、df_districts
    回傳 (csrf, captcha_key, df_districts, response_query)
    """
    # Step 1: 進主頁（讓 session 拿到 cookie）
    session.get("https://www.ris.gov.tw/info-doorplate/app/doorplate/main")
 
    # Step 2: POST map 頁（searchType=date）
    response_map = session.post(
        "https://www.ris.gov.tw/info-doorplate/app/doorplate/map",
        data={"searchType": "date"}
    )
 
    # Step 3: 抓 CSRF（先找 input，找不到找 meta，都沒有就用空字串）
    soup_map = BeautifulSoup(response_map.text, "html.parser")
    csrf_input = soup_map.find("input", {"name": "_csrf"})
    csrf_meta  = soup_map.find("meta",  {"name": "_csrf"})
 
    if csrf_input:
        csrf = csrf_input["value"]
    elif csrf_meta:
        csrf = csrf_meta["content"]
    else:
        csrf = ""   # 有些 session 不需要 csrf 也能過
        print("⚠️  找不到 CSRF，用空字串繼續（通常不影響）")
 
    print(f"csrf = {csrf!r}")
 
    # Step 4: POST query 頁（選台北市）
    response_query = session.post(
        "https://www.ris.gov.tw/info-doorplate/app/doorplate/query",
        data={
            "searchType": "date",
            "cityCode":   CITY_CODE,
            "_csrf":      csrf
        }
    )
 
    soup_query = BeautifulSoup(response_query.text, "html.parser")
```

```python
# ── Cell 3：顯示驗證碼 ────────────────────────────────────
def show_captcha(captcha_key: str) -> bytes:
    """下載驗證碼圖片並在 Jupyter 顯示，回傳圖片 bytes"""
    url = (
        "https://www.ris.gov.tw/info-doorplate/captcha/image"
        f"?CAPTCHA_KEY={captcha_key}&time={int(time.time()*1000)}"
    )
    img_bytes = session.get(url).content
    with open("captcha.jpg", "wb") as f:
        f.write(img_bytes)
    display(Image(filename="captcha.jpg"))
    return img_bytes
 
show_captcha(captcha_key)
```

```python
# ── Cell 7：全區主流程 ────────────────────────────────────
def run_all_districts(df_districts, csrf, captcha_key):
    """
    逐區查詢，每區需人工輸入驗證碼。
    查完自動從回應取得下一個 captchaKey。
    """
    current_key = captcha_key
    results     = []
 
    for i, row in df_districts.iterrows():
        area_name = row["區名"]
        area_code = row["區代碼"]
 
        print(f"\n{'='*50}")
        print(f"[{i+1}/{len(df_districts)}] 查詢：{area_name}（{area_code}）")
 
        # 顯示驗證碼
        show_captcha(current_key)
 
        # 人工輸入
        captcha_input = input(f"請輸入「{area_name}」驗證碼：").strip()
 
        # 查詢
        data = query_area(
            area_code     = area_code,
            captcha_input = captcha_input,
            captcha_key   = current_key,
            csrf          = csrf,
        )
 
        record_count = data.get("records", 0)
 
        if record_count == 0:
            print(f"❌ {area_name} 查詢失敗（驗證碼錯誤或無資料）")
            # 重新拿驗證碼再試一次
            retry = input("重試？(y/n)：").strip().lower()
            if retry == "y":
                # captchaKey 從失敗回應裡拿新的
                next_key = get_next_captcha(data)
                if next_key:
                    current_key = next_key
                show_captcha(current_key)
                captcha_input = input(f"重新輸入「{area_name}」驗證碼：").strip()
                data = query_area(area_code, captcha_input, current_key, csrf)
                record_count = data.get("records", 0)
                if record_count == 0:
                    print(f"❌ {area_name} 再次失敗，跳過")
                    results.append({"區名": area_name, "狀態": "失敗", "筆數": 0})
                    next_key = get_next_captcha(data)
                    if next_key:
                        current_key = next_key
                    continue
 
        # 儲存（CSV + DB）
        count = save_area(area_name, data)
        results.append({"區名": area_name, "狀態": "成功", "筆數": count})
        print(f"✅ {area_name} 完成，共 {count} 筆")
 
        # 取下一個 captchaKey
        next_key = get_next_captcha(data)
        if next_key:
            current_key = next_key
        else:
            print("⚠️  無法取得下一個 captchaKey，重新初始化 session")
            csrf, current_key, _ = setup_and_get_districts()
 
        time.sleep(1)   # 避免太快被封
 
    print(f"\n{'='*50}")
    print("全部完成！結果摘要：")
    print(pd.DataFrame(results).to_string(index=False))
    return results
```

```python
# ── Cell 8：執行！──────────────────────────────────────────
# 確認 df_districts、csrf、captcha_key 都已取得後執行
results = run_all_districts(df_districts, csrf, captcha_key)
 
```

```python
from bs4 import BeautifulSoup
import pandas as pd

soup = BeautifulSoup(response_query.text, "html.parser")

area_select = soup.find("select", {"name": "areaCode"})

districts = []

for opt in area_select.find_all("option"):

    code = opt.get("value", "").strip()
    name = opt.get_text(strip=True)

    if code and "選擇" not in name:
        districts.append({
            "區名": name,
            "區代碼": code
        })

df_districts = pd.DataFrame(districts)

print(f"共抓到 {len(df_districts)} 個行政區")

df_districts
```

```python
# 建立半自動流程
def query_area_manual(
    area_name,
    area_code,
    captcha_key
):

    print("="*50)
    print(f"查詢 {area_name}")

    show_captcha(captcha_key)

    captcha_input = input(
        f"{area_name} 驗證碼："
    )

    data = query_area(
        area_code,
        captcha_input=captcha_input,
        captcha_key=captcha_key
    )

    if data.get("records", 0) == 0:

        print("❌ 查詢失敗")

        return None, captcha_key

    print(
        f"✅ 成功 {data['records']} 筆"
    )

    save_area_csv(area_name, data)

    next_key = get_next_captcha(data)

    return data, next_key
```

```python
query_api = "https://www.ris.gov.tw/info-doorplate/app/doorplate/inquiry/date"

def query_area(area_code):

    payload = {
        "searchType": "date",
        "cityCode": "63000000",
        "tkt": "-1",
        "areaCode": area_code,
        "village": "",
        "neighbor": "",
        "sDate": "114-09-01",
        "eDate": "114-11-30",
        "_includeNoDate": "on",
        "registerKind": "1",
        "captchaInput": captcha_text,
        "captchaKey": captcha_key,
        "_csrf": current_csrf,
        "_search": "false",
        "rows": "9999",
        "page": "1",
        "sord": "asc"
    }

    r = session.post(query_api, data=payload)
    return r.json()
```

```python
import json

def get_next_captcha(data):

    try:

        msg = json.loads(data["errorMsg"])

        return msg.get("captcha")

    except Exception as e:

        print("captcha 解析失敗")

        return None
```

```python
# 全行政區抓取

all_rows = []

for _, row in df_districts.iterrows():

    area_name = row["區名"]
    area_code = row["區代碼"]

    data = query_area(area_code)

    rows = data.get("rows", [])

    for r in rows:
        r["行政區"] = area_name

    all_rows.extend(rows)

print("總筆數:", len(all_rows))
```

```python
register_kind_map = {
    "1": "門牌初編",
    "2": "門牌增編",
    "3": "門牌改編",
    "4": "門牌合併",
    "5": "門牌廢止",
    "6": "行政區域調整",
    "7": "門牌整編"
}

final_data_list = []

for row in all_rows:

    final_data_list.append({
        "行政區": row["行政區"],
        "門牌資料": row["v1"],
        "編訂日期": row["v2"],
        "編訂類別": row["v3"]
    })
```

```python
def clean_address(full_address):

    pattern = re.compile(
        r'(?P<city>臺北市|台北市)'
        r'(?P<district>[^區]+區)'
        r'(?P<village>[^里]+里)?'
        r'(?P<neighborhood>\d+鄰)?'
        r'(?P<detail>.*)'
    )

    match = pattern.match(full_address)

    if match:
        data = match.groupdict()

        if not data['village']:
            data['village'] = "未分類"

        if not data['neighborhood']:
            data['neighborhood'] = "未分類"

        return data

    return {
        "city":"臺北市",
        "district":"未知區",
        "village":"未分類",
        "neighborhood":"未分類",
        "detail":full_address
    }

def clean_date(roc_date_str):

    nums = re.findall(r'\d+', roc_date_str)

    if len(nums) == 3:

        year = int(nums[0]) + 1911
        month = int(nums[1])
        day = int(nums[2])

        return f"{year}-{month:02d}-{day:02d}"

    return roc_date_str
```

```python
def process_and_clean_data(final_data_list):

    df = pd.DataFrame(final_data_list)

    df["編釘日期"] = df["編釘日期"].apply(clean_date)

    address_df = pd.json_normalize(
        df["門牌資料"].apply(clean_address)
    )

    df = df.rename(
        columns={"門牌資料":"原始門牌地址"}
    )

    cleaned_df = pd.concat(
        [address_df, df],
        axis=1
    )

    return cleaned_df

cleaned_df = process_and_clean_data(
    final_data_list
)

cleaned_df.head()
```

```python
import pandas as pd

def save_area_csv(area_name, data):

    rows = data.get("rows", [])

    if not rows:
        print(f"❌ {area_name} 沒有資料")
        return

    result = []

    for row in rows:

        result.append({
            "行政區": area_name,
            "門牌資料": row["v1"],
            "編訂日期": row["v2"],
            "編訂類別": row["v3"]
        })

    df = pd.DataFrame(result)

    filename = f"{area_name}.csv"

    df.to_csv(
        filename,
        index=False,
        encoding="utf-8-sig"
    )

    print(f"✅ 已儲存 {filename}")
```
